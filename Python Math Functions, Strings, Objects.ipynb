# `float` vs `decimal`

* `float` is used for scientific calculations - performance > precision 
* `decimal` is used in accounting/finance - precision > performance

```python
float(1)
```

```python
type(1.0)
```

```python
import decimal
```

`decimal` is a number represented using a decimal point, separating the whole number part from the fractional(decimal) part. 

In Python, floating point is always represented in approximation - due to binary fractions, there will be small rounding errors: 

```python
decimal.Decimal(3.45)
```

```python
decimal.Decimal(3.44)
```

To declare a precise number, use the single quotes: 

```python
decimal.Decimal('3.45')
```

`Banker's Rounding` - round-half-to-even - applies **only** when the number is half-way between two integers 

```python
round(6.5)
```

```python
round(decimal.Decimal('6.5'))
```

<div style="background-color:#DBEAFE; color:#1E40AF; padding:10px; border-radius:5px;">
Python's built-in <code>round()</code> function uses <strong>Banker's rounding</strong>, but rounding floats can lead to unexpected results due to binary imprecision.<br>
<br>
Always use <code>decimal.Decimal</code> constructed from strings (e.g., <code>Decimal('5.45')</code>) for predictable rounding.
</div>

# `ascii` and `unicode`

`character encoding` - mapping a character to its binary representation

`ASCII` - American Standard Code for Information Interchange:
* 8-bit encoding scheme: uses numbers 0-127 to enumerate characters 
* upper and lowercase letters
* digits
* punctuation marks 
* control characters 

`Unicode` - encoding scheme for representing international characters. `ASCII` is a subset of `Unicode`.
* `\u` + 4 hexademical digits: from `\u0000` to `\uFFFF`

| Characters  | ASCII | Unicode       |
|-------------|-----------------------|---------------------|
| `'0'`–`'9'` | 48–57                | `\u0030`–`\u0039`   |
| `'A'`–`'Z'` | 65–90                | `\u0041`–`\u005A`   |
| `'a'`–`'z'` | 97–122               | `\u0061`–`\u007A`   |

<div style="background-color:#FAF9F6; color:#283618; padding:10px; border-radius:5px;">
<b>Assignment</b>: Write a program that prompts the user to enter a letter grade A/a, B/b, C/c, D/d, or F/f and displays its corresponding numeric value 4, 3, 2, 1, or 0.

Sample Run 1

Enter a letter grade: B
The numeric value for grade B is 3
Sample Run 2

Enter a letter grade: b
The numeric value for grade b is 3
Sample Run 3

Enter a letter grade: T
T is an invalid grade
</div>

```python
grade = input("Enter a letter grade: ").strip()

# Convert the grade to uppercase so we can handle both upper/lowercase uniformly
grade_upper = grade.upper()

# Map each valid grade to its numeric value
if grade_upper == 'A':
    numeric_value = 4
elif grade_upper == 'B':
    numeric_value = 3
elif grade_upper == 'C':
    numeric_value = 2
elif grade_upper == 'D':
    numeric_value = 1
elif grade_upper == 'F':
    numeric_value = 0
else:
    print(f"{grade} is an invalid grade")

print(f"The numeric value for grade {grade} is {numeric_value}")
```
